## 학습

```python
import urllib.request
from soynlp import DoublespaceLineCorpus
from soynlp.word import WordExtractor
```

```python
urllib.request.urlretrieve("https://raw.githubusercontent.com/lovit/soynlp/master/tutorials/2016-10-20.txt", filename="2016-10-20.txt")
```

```python
corpus = DoublespaceLineCorpus('2016-10-20.txt')
len(corpus)
```

```python
word_extractor = WordExtractor()
word_extractor.train(corpus)
```

```python
word_score_table = word_extractor.extract()
```

```python
word_score_table
```

```python
word_score_table['반포한강공'].cohesion_forward
```

```python
word_score_table['반포한강공원'].cohesion_forward
```

```python
word_score_table['반포한강공원에'].cohesion_forward
```

```python
word_score_table['반포한강공'].right_branching_entropy
```

```python
word_score_table['반포한강공원'].right_branching_entropy
```

```python
scores = {word:score.cohesion_forward for word, score in word_score_table.items()}
scores # 단어:응집 score
```

## Tokenize

```python
from soynlp.tokenizer import LTokenizer

ltokenizer = LTokenizer(scores=scores)

print(ltokenizer.tokenize('국제사회와 우리의 노력들로 범죄를 척결하자.', flatten=False))
print(ltokenizer.tokenize('국제사회와우리의노력들로범죄를척결하자.'))
```

```python
from soynlp.tokenizer import MaxScoreTokenizer

maxscore_tokenizer = MaxScoreTokenizer(scores=scores)

print(maxscore_tokenizer.tokenize('국제사회와우리의노력들로범죄를척결하자.'))
print(maxscore_tokenizer.tokenize('국제사회와 우리의 노력들로 범죄를 척결하자.'))
```
